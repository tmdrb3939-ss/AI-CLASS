```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/breast_cancer.csv')

y = df['worst compactness']
y
```

```python
X = df.drop(['worst compactness', 'label', 'Unnamed: 0'], axis=1, errors='ignore')
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

reg_lr = LinearRegression()
reg_lr.fit(X_train, y_train)

pred_lr = reg_lr.predict(X_test)

print("\n--- [Breast Cancer] Linear Regression (Worst Compactness) ---")
print("RMSE:", np.sqrt(mean_squared_error(y_test, pred_lr)))
print("R2 Score:", r2_score(y_test, pred_lr))
```
